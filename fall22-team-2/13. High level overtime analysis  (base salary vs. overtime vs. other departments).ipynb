```python
import random

def roundFloat(x):
    if x - int(x) >= .5:
        return int(x)+1
    return int(x)+.5

def int_to_str(x):
    num ='$'
    if x == 0:
        num += '0'
    elif 0 < x < 999:
        num+=str(x)
    elif 999 < x < 999999:
        num += "{:,}".format(int(x))
    elif 999999 < x < 999999999:
        if x / 1000000 > 99 or int(x / 1000000) == (x / 1000000):
            num += str(int(x / 1000000)) + ' mil'
        else:
            num += str((x / 1000000)) + ' mil'
    elif 999999999 < x < 999999999999:
        if x / 1000000000 > 99 or int(x / 1000000000) == (x / 1000000000):
            num += str(int(x / 1000000000)) + ' bil'
        else:
            num += str((x / 1000000000)) + ' bil'
    elif 999999999999 < x < 999999999999999:
        if x / 1000000000000 > 99 or int(x / 1000000000000) == (x / 1000000000000):
            num += str(int(x / 1000000000000)) + ' tril'
        else:
            num += str((x / 1000000000000)) + ' tril'
    return num+' '


def ls_to_string(ls):
    ls_str = []
    for i in ls:
        ls_str.append(int_to_str(i))
    return ls_str

def custom_ticks(maxY):
    degree = 1
    counter = maxY
    while counter/10 > 1:
        degree*=10
        counter /= 10

    tick =degree*roundFloat(counter)/5
    ls = [0]
    for i in range(1,7):
        ls.append(i * tick)

    return (ls, ls_to_string(ls))

for i in range(145,155):
    v= random.randint(1,99999999)
    print("________  ",v,"  ________")

    print("     ", custom_ticks(v))
```
