```python
earnings21 = pd.DataFrame(pd.read_csv('./Data/employees-2021.csv', encoding='latin1'))
earnings21['TOTAL_GROSS']=earnings21['TOTAL_GROSS'].astype(str)
earnings21['TOTAL_GROSS']=earnings21['TOTAL_GROSS'].apply(lambda x: float(x.replace(",","")))

earnings22 = pd.DataFrame(pd.read_csv('./Data/employees-2022.csv', encoding='latin1'))
earnings22['TOTAL_ GROSS']=earnings22['TOTAL_ GROSS'].astype(str)
earnings22['TOTAL_ GROSS']=earnings22['TOTAL_ GROSS'].apply(lambda x: float(x.replace(",","")))


fpath = Path("./fonts/ttf/GT-America-Standard-Regular.ttf")
fpath_bold = Path("./fonts/ttf/GT-America-Standard-Bold.ttf")
```

```python
bpd_earnings21 = earnings21[earnings21['DEPARTMENT_NAME']=='Boston Police Department']
nonbpd_earnings21 = earnings21[earnings21['DEPARTMENT_NAME']!='Boston Police Department']
colors = ["#FCAA17", "#FCAA17", "#EF404E",'#552564','#0055AA',
    '#130F54','#3C3532','#691B40','#231F20','#EFECEA',
    '#FFE06A','#FFE06A','#FABEAF','#A3DBE3','#CCCCCC' ]
font = fpath


# make data
x = ['BPD', 'Remainder of City Budget']
y = [bpd_earnings21['TOTAL_GROSS'].sum(),
     nonbpd_earnings21['TOTAL_GROSS'].sum()]
# plot
fig, ax = plt.subplots()

fig.set_figheight(10)
fig.set_figwidth(9)


cont = ax.bar(x,y, color=['#EF404E',"#BBBBBB"])
ax.set(ylim=(0, 1600000000),yticks=range(0,1600000000,300000000))
plt.ticklabel_format(style='plain', axis='y', useLocale=True)
#plt.xlabel('Department', fontsize=16, font=font)
#plt.ylabel('Total Payroll', fontsize=16, font=font)
#plt.title('Boston Police Department Budget Makes up Nearly a Quarter of the Cities Budget',fontsize=18, font=font)
ax.set_title('Boston Police Department budget made up nearly a\nquarter of the city budget in 2021', font=fpath_bold, fontsize=21, loc='left', pad=40)
ax.title.set_size(20)


plt.grid(which='major', axis='y', color='lightgrey', zorder=10.0)
ax.set_axisbelow(True)

ax.set_yticklabels(['$0 ','$300 mil ','$600 mil ','$900 mil ','$1.2 bil ','$1.5 bil '])


locale.setlocale(locale.LC_ALL, 'en_CA.UTF-8')
labels = []
for i in y:
    labels.append('   ${:,.0f}'.format(i))
    
# ax.bar_label(container=cont, labels=labels,fontsize=14,font=font)

plt.rcParams.update({'font.size': 15})
plt.xticks(font=fpath)
plt.yticks(font=fpath)
for pos in ['right', 'top', 'bottom', 'left']:
   plt.gca().spines[pos].set_visible(False)
   
fig.savefig('./Hayden_Images/8/Boston Police Department budget makes up nearly a quarter of the cities budget.png')



plt.show()
```

```python
bpd_earnings22 = earnings22[earnings22['DEPARTMENT_NAME']=='Boston Police Department']
nonbpd_earnings22 = earnings22[earnings22['DEPARTMENT_NAME']!='Boston Police Department']
colors = ["#FCAA17", "#FCAA17", "#EF404E",'#552564','#0055AA',
    '#130F54','#3C3532','#691B40','#231F20','#EFECEA',
    '#FFE06A','#FFE06A','#FABEAF','#A3DBE3','#CCCCCC' ]
font = fpath


# make data
x = ['Other City Employees','BPD Employees']
y = [ nonbpd_earnings22['TOTAL_ GROSS'].sum() 
     / len(nonbpd_earnings22['TOTAL_ GROSS'].index),
     bpd_earnings22['TOTAL_ GROSS'].sum() 
     / len(bpd_earnings22['TOTAL_ GROSS'].index)]
# plot
fig, ax = plt.subplots()

fig.set_figheight(10)
fig.set_figwidth(10)


cont = ax.bar(x,y, color=["#BBBBBB",'#EF404E'])
ax.set(ylim=(0, 160000),yticks=range(0,160000,30000))
plt.ticklabel_format(style='plain', axis='y', useLocale=True)

ax.set_title('On average, BPD employees\' salaries were nearly double that of \nother city employees in 2022',font=fpath_bold, fontsize=20, loc='left', pad=40)
ax.set_title('The average BPD salary was 77% more than the \naverage salary of other city employees in 2022 ',font=fpath_bold, fontsize=20, loc='left', pad=40)


locale.setlocale(locale.LC_ALL, 'en_CA.UTF-8')
labels = []
for i in y:
    labels.append('   ${:,.0f}'.format(i))


fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
ax.bar_label(container=cont, labels=labels,fontsize=14,font=font)
plt.xticks(font=fpath)
plt.yticks(font=fpath)
#ax.yaxis.set_major_formatter('${x:1.0f}')

plt.grid(which='major', axis='y', color='lightgrey', zorder=10.0)
ax.set_axisbelow(True)
plt.tick_params(left = False, bottom=False)




for pos in ['right', 'top', 'bottom', 'left']:
   plt.gca().spines[pos].set_visible(False)
fig.savefig('./Hayden_Images/8/On average, BPD employees salaries are nearly double that of all other city employees.png')

plt.show()
```
