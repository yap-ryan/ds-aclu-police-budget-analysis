```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the xlsx file and spreadsheet 

df2012= pd.read_excel('./Data/Special_Events_Overtime_2012-2021/Special_Events_Overtime_2012-2022.xlsx',sheet_name='2012')
df2013= pd.read_excel('./Data/Special_Events_Overtime_2012-2021/Special_Events_Overtime_2012-2022.xlsx',sheet_name='2013')
df2014= pd.read_excel('./Data/Special_Events_Overtime_2012-2021/Special_Events_Overtime_2012-2022.xlsx',sheet_name='2014')
df2015= pd.read_excel('./Data/Special_Events_Overtime_2012-2021/Special_Events_Overtime_2012-2022.xlsx',sheet_name='2015')
df2016= pd.read_excel('./Data/Special_Events_Overtime_2012-2021/Special_Events_Overtime_2012-2022.xlsx',sheet_name='2016')
df2017= pd.read_excel('./Data/Special_Events_Overtime_2012-2021/Special_Events_Overtime_2012-2022.xlsx',sheet_name='2017')
df2018= pd.read_excel('./Data/Special_Events_Overtime_2012-2021/Special_Events_Overtime_2012-2022.xlsx',sheet_name='2018')
df2019= pd.read_excel('./Data/Special_Events_Overtime_2012-2021/Special_Events_Overtime_2012-2022.xlsx',sheet_name='2019')
df2020= pd.read_excel('./Data/Special_Events_Overtime_2012-2021/Special_Events_Overtime_2012-2022.xlsx',sheet_name='2020')
df2021= pd.read_excel('./Data/Special_Events_Overtime_2012-2021/Special_Events_Overtime_2012-2022.xlsx',sheet_name='2021')
df2022= pd.read_excel('./Data/Special_Events_Overtime_2012-2021/Special_Events_Overtime_2012-2022.xlsx',sheet_name='2022')


# extract the DESCRIPTION column that has MASS / MELNEA 
df2020mass = df2020[df2020['DESCRIPTION'].str.contains('MASS')]
df2021mass= df2021[df2021['DESCRIPTION'].str.contains('MASS')]
df2022mass = df2022[df2022['DESCRIPTION'].str.contains('MASS')]
```

```python
# Distribution density graph of the OTHOURS from df2012 to df2019 in subplots, Y axis is the density of the OTHOURS, X axis is the OTHOURS
fig, axs = plt.subplots(4, 2, figsize=(15, 15))

axs[0, 0].set_title('2012')
axs[0, 1].set_title('2013')
axs[1, 0].set_title('2014')
axs[1, 1].set_title('2015')
axs[2, 0].set_title('2016')
axs[2, 1].set_title('2017')
axs[3, 0].set_title('2018')
axs[3, 1].set_title('2019')

fig.tight_layout(pad=3.0)

axs[0, 0].grid(axis='y')
axs[0, 1].grid(axis='y')
axs[1, 0].grid(axis='y')
axs[1, 1].grid(axis='y')
axs[2, 0].grid(axis='y')
axs[2, 1].grid(axis='y')
axs[3, 0].grid(axis='y')
axs[3, 1].grid(axis='y')

sns.distplot(df2012['OTHOURS'], ax=axs[0, 0])
sns.distplot(df2013['OTHOURS'], ax=axs[0, 1])
sns.distplot(df2014['OTHOURS'], ax=axs[1, 0])
sns.distplot(df2015['OTHOURS'], ax=axs[1, 1])
sns.distplot(df2016['OTHOURS'], ax=axs[2, 0])
sns.distplot(df2017['OTHOURS'], ax=axs[2, 1])
sns.distplot(df2018['OTHOURS'], ax=axs[3, 0])
sns.distplot(df2019['OTHOURS'], ax=axs[3, 1])

plt.show()



# Distribution density graph in fold line graph of the OTHOURS from df2012 to df2019 in subplots, Y axis is the density of the OTHOURS, X axis is the OTHOURS




```

```python
# Do the thing above cell by using plotly package
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "notebook_connected"


fig = go.Figure()

fig.add_trace(go.Histogram(x=df2012['OTHOURS'], name='2012'))

# set the x axis
fig.update_xaxes(title_text='OTHOURS')
# set the y axis
fig.update_yaxes(title_text='Number of OTHOURS')

# set the title
fig.update_layout(title_text='<b> Overtime hours distribution of year 2012<b>')

fig.update_layout(font=dict(family="GT America"))
fig.update_layout(font_family="GT America Bold")






```
