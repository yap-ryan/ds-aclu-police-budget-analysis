# IMPORT PACKAGE

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# IMPORT DATA

```python
#read the xlsx file and spreadsheet 

df2012 = pd.read_excel('Special_Events_Overtime_2012-2022.xlsx',sheet_name='2012')
df2013= pd.read_excel('Special_Events_Overtime_2012-2022.xlsx',sheet_name='2013')
df2014= pd.read_excel('Special_Events_Overtime_2012-2022.xlsx',sheet_name='2014')
df2015= pd.read_excel('Special_Events_Overtime_2012-2022.xlsx',sheet_name='2015')
df2016= pd.read_excel('Special_Events_Overtime_2012-2022.xlsx',sheet_name='2016')
df2017= pd.read_excel('Special_Events_Overtime_2012-2022.xlsx',sheet_name='2017')
df2018= pd.read_excel('Special_Events_Overtime_2012-2022.xlsx',sheet_name='2018')
df2019= pd.read_excel('Special_Events_Overtime_2012-2022.xlsx',sheet_name='2019')
df2020= pd.read_excel('Special_Events_Overtime_2012-2022.xlsx',sheet_name='2020')
df2021= pd.read_excel('Special_Events_Overtime_2012-2022.xlsx',sheet_name='2021')
df2022= pd.read_excel('Special_Events_Overtime_2012-2022.xlsx',sheet_name='2022')


```

```python

# extract the DESCRIPTION column that has MASS / MELNEA 
df2020mass = df2020[df2020['DESCRIPTION'].str.contains('MASS')]
df2021mass= df2021[df2021['DESCRIPTION'].str.contains('MASS')]
df2022mass = df2022[df2022['DESCRIPTION'].str.contains('MASS')]
```

```python
# Distribution density graph of the OTHOURS from df2012 to df2019 in subplots, Y axis is the density of the OTHOURS, X axis is the OTHOURS
fig, axs = plt.subplots(4, 2, figsize=(15, 15))

axs[0, 0].set_title('2012')
axs[0, 1].set_title('2013')
axs[1, 0].set_title('2014')
axs[1, 1].set_title('2015')
axs[2, 0].set_title('2016')
axs[2, 1].set_title('2017')
axs[3, 0].set_title('2018')
axs[3, 1].set_title('2019')

fig.tight_layout(pad=3.0)

axs[0, 0].grid(axis='y')
axs[0, 1].grid(axis='y')
axs[1, 0].grid(axis='y')
axs[1, 1].grid(axis='y')
axs[2, 0].grid(axis='y')
axs[2, 1].grid(axis='y')
axs[3, 0].grid(axis='y')
axs[3, 1].grid(axis='y')

sns.distplot(df2012['OTHOURS'], ax=axs[0, 0])
sns.distplot(df2013['OTHOURS'], ax=axs[0, 1])
sns.distplot(df2014['OTHOURS'], ax=axs[1, 0])
sns.distplot(df2015['OTHOURS'], ax=axs[1, 1])
sns.distplot(df2016['OTHOURS'], ax=axs[2, 0])
sns.distplot(df2017['OTHOURS'], ax=axs[2, 1])
sns.distplot(df2018['OTHOURS'], ax=axs[3, 0])
sns.distplot(df2019['OTHOURS'], ax=axs[3, 1])

plt.show()



# Distribution density graph in fold line graph of the OTHOURS from df2012 to df2019 in subplots, Y axis is the density of the OTHOURS, X axis is the OTHOURS




```

```python
# did the same thing above for df2020mass to df2022mass
fig, axs = plt.subplots(3, 1, figsize=(15, 15))

axs[0].set_title('2020')
axs[1].set_title('2021')
axs[2].set_title('2022')

fig.tight_layout(pad=3.0)

axs[0].grid(axis='y')
axs[1].grid(axis='y')
axs[2].grid(axis='y')

sns.distplot(df2020mass['OTHOURS'],
                ax=axs[0])
sns.distplot(df2021mass['OTHOURS'],
                ax=axs[1])
sns.distplot(df2022mass['OTHOURS'],
                ax=axs[2])

plt.show()
```
