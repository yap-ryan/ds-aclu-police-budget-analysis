```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px

# file=pd.read_csv('preYear.csv')
# set(file)
file21=pd.read_csv('./Data/employees-2021.csv',thousands=',', encoding='latin-1')
file20=pd.read_csv('./Data/employees-2020.csv',thousands=',', encoding='latin-1')
file19=pd.read_csv('./Data/employees-2019.csv',thousands=',', encoding='latin-1')
file18=pd.read_csv('./Data/employees-2018.csv',thousands=',', encoding='latin-1')
file17=pd.read_csv('./Data/employees-2017.csv',thousands=',', encoding='latin-1')
file16=pd.read_csv('./Data/employees-2016.csv',thousands=',', encoding='latin-1')
file15=pd.read_csv('./Data/employees-2015.csv',thousands=',', encoding='latin-1')
file14=pd.read_csv('./Data/employees-2014.csv',thousands=',', encoding='latin-1')
file13=pd.read_csv('./Data/employees-2013.csv',thousands=',', encoding='latin-1')
file12=pd.read_csv('./Data/employees-2012.csv',thousands=',', encoding='latin-1')
file11=pd.read_csv('./Data/employees-2011.csv',thousands=',', encoding='latin-1')
```

```python
df21 = pd.DataFrame(file21)
top21=df21.sort_values(by='TOTAL_GROSS',ascending=False)

top21_20=top21.head(20)
label2021=list(top21_20['NAME'])
value=list(top21_20['TOTAL_GROSS'])
plt.rcdefaults()
fig, ax = plt.subplots()
people=label2021
y_pos = np.arange(len(people))
performance=value
ax.barh(y_pos, performance)
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()# labels read top-to-bottom
ax.set_yticklabels(people, fontsize=10)
ax.set_xlabel('unit=100')
ax.set_title('2021 Top Earners')
#Change the color or bars to red and black
ax.barh(y_pos, performance, color=['red', 'black'])
# Add the gridlines
ax.grid(b=True, color='grey', linestyle='-.', linewidth=0.5, alpha=0.2)

plt.show()
```
