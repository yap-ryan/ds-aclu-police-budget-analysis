## We will the data form 2021 to analysis.
Data from https://data.boston.gov/dataset/employee-earnings-report.

```python
import pandas as pd
df=pd.read_csv('./Data/employees-2022.csv',thousands=',', encoding='latin-1')
# df.head(10)
# type(df['OTHER'][0])
non_bpd_df = df[df.DEPARTMENT_NAME != 'Boston Police Department']
non_bpd_df.head(10)
```

```python
import plotly.graph_objects as go

df = df.loc[df['DEPARTMENT_NAME'] == 'Boston Police Department']

total_regular=df['REGULAR'].sum()
total_retro=df['RETRO'].sum()
total_other=df['OTHER'].sum()
total_detail=df['DETAIL'].sum()
total_gross=df['TOTAL_ GROSS'].sum()
total_overtime=df['OVERTIME'].sum()
total_quinn=df['QUINN_EDUCATION'].sum()
total_injured=df['INJURED'].sum()
data=[total_regular,total_retro,total_other,total_overtime,total_quinn,total_injured, total_detail]
names=['Regular','Retro','Other','Overtime','Quinn Education Incentive','Injured','Detail']


import matplotlib.pyplot as plt
import plotly.express as px

fig = px.pie(df, values=data, names=names, color_discrete_sequence=["#130F54",'#EF404E','#BBBBBB', '#FCAA17', '#691B40', '#0055AA', '#EFECEA'],)
fig.update_layout(font_family="GT America Bold")
# change the frount of thing other than title to GT-America-Standard-regular

fig.update_layout(
    font_family="GT America",   
    font_size=12, 
    width=630,
    height=640)

fig.update_layout(
    title = go.layout.Title(
        text = "Type Distribution of BPD Payroll in 2022<br>",
        xref = "paper",
        x = 0.5,
        font_family = "GT America Bold",
        font_size = 20,
        font_color="black"
    ),
    margin=dict(
        t=100
    )
)

fig.update_traces(showlegend=False, textinfo='percent+label')

fig.show()
print(data)
```

```python
import seaborn as sns

sns.set_style('whitegrid')
sns.distplot(df['REGULAR'], kde = True, color ='#0055aa', bins = 30)
# bold the title
#remove the x and y labels
plt.xlabel('')
plt.ylabel('')
# change the title to GT America format
plt.title('Distribution of Regular Pay', fontweight ="bold", fontfamily = 'GT America Bold')
#  Remove the border line box around plots
sns.despine(left = True, bottom = True)
plt.show()
```

```python
# use plotly to do the thing in above cell
import math
import plotly.express as px
from plotly.graph_objs import *

# Remove negative outliers
df = df.drop(df.index[df['REGULAR'] <= 0])
non_bpd_df = non_bpd_df.drop(non_bpd_df.index[non_bpd_df['REGULAR'] <= 0])
non_bpd_df['REGULAR']=non_bpd_df['REGULAR'].dropna()


layout = Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)

# fig.update_traces(marker_color=['#0055aa'])

# make the background transparent in white
fig.update_layout(plot_bgcolor='rgba(0,0,0,0)')
fig = px.histogram(df, x="REGULAR", nbins=50, title='<b>Distribution of Regular Pay</b>', color_discrete_sequence=['#130F54'])

# remove the x and y labels
fig.update_yaxes(title_text='Number of employmees')
fig.update_yaxes(showgrid=True, gridwidth=0.5, gridcolor='LightGray')

# Median line
print(df['REGULAR'].median())
print(non_bpd_df['REGULAR'].median())

fig.add_vline(x=df['REGULAR'].median(), 
    line_width=2, 
    line_color="red", 
    # annotation_text="BPD Median: $" + str(math.trunc(df['REGULAR'].median())), 
    annotation_text="BPD Median", 
    annotation_position="left top",
    # annotation=dict(font_size=13, font_family="GT America", xref="x domain", x=0.36, y=1.02),
    annotation=dict(font_size=14, font_color="grey", font_family="GT America", textangle=-90)

)

fig.add_vline(x=non_bpd_df['REGULAR'].median(), 
    line_width=2, 
    line_color="red",
    line_dash="dash",  
    # annotation_text="Other Depts Median: $" + str(math.trunc(non_bpd_df['REGULAR'].median())), 
    annotation_text="Other Depts Median", 
    annotation_position="left top",
    # annotation=dict(font_size=13, font_family="GT America", xref="x domain", x=0.16, y=1.02),
    annotation=dict(font_size=14, font_color="grey",font_family="GT America", textangle=-90)

)

# Font & other aesthetic stuff
fig.update_layout(
    plot_bgcolor='rgba(0,0,0,0)',
    font_family="GT America",
    font_color="black",
    xaxis_title=None,
    xaxis_tickprefix = '$',
    height=500,
    width=1100,
    margin=dict(
        l=100,
        r=100,
        b=80,
        t=125,
        pad=4
    )
    
)

# Customize title
fig.update_layout(
    title=go.layout.Title(
        text="Distribution of BPD's Regular Pay in 2022",
        xref="paper",
        x=0,
        y=0.9,
        font_family="GT America Bold",
        font_size=19
    )
)

fig.show()
```
